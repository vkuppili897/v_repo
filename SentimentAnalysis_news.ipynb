```
import pandas as pd
import io
from datetime import datetime
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
import nltk
nltk.download('punkt')
nltk.download('movie_reviews')
```

```
df = pd.read_csv('CombinedDated.csv',encoding='latin1')
df = df.filter(items=['Date','Company','Publisher','Headline', 'Article'],axis=1)
df = df.set_index(['Date'])
print(df)
```

```
print(df.describe())
print(df.dtypes,df.columns)
```

```
def TextAnalyser(text):

  sent = TextBlob(text)
  polarity = sent.sentiment.polarity
  subjectivity = sent.sentiment.subjectivity
  sent = TextBlob(text, analyzer = NaiveBayesAnalyzer())
  classification = sent.sentiment.classification
  positive = sent.sentiment.p_pos
  negative = sent.sentiment.p_neg

  print(classification, positive, negative)

  return(classification, positive, negative)

t = ("Vivek is a great cook")
TextAnalyser(t)
```

```
lines = df['Article']

lst = []

for l in lines:
  x = TextAnalyser(l)
  lst.append(x)

```

```
def filter_values(lst):
  new_lst = []

  for tuple in lst:
    (sentiment, pol, sub) = tuple
    if sentiment == 'pos':
      new_lst.append((1, pol, sub))
    else: 
      new_lst.append((0, pol, sub))
  
  return new_lst

result = pd.DataFrame(filter_values(lst))
print(result)
```
